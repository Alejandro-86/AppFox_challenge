```python
from collections import Counter
'''
Returns the elements that occur i times within an array.
Inputs: Array-like and i.
output: list of unique numbers or empty list.
'''
def elements_n_occurrences(array, i):
    # Error cases
    if not array: return []
    if len(array)<2: return []
    # -----------
    
    # storing count of elements in cont_two
    cont_two = Counter(array)
    
    # list comprehension with the elements that have n occurences
    return [n for n in cont_two.keys() if cont_two[n] == i]

    
x = [1,1,4,2,2]
i = 2
elements_n_occurrences(x, i)
```

```python
import tkinter as tk
import urllib.request, json 
import random
import matplotlib.pyplot as plt
from PIL import Image
from urllib.request import urlopen
from IPython.display import clear_output


class Application(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.pack()
        self.create_widgets()

    def create_widgets(self):
        self.hi_there = tk.Button(self)
        self.hi_there["text"] = "Create a character\n (click me)"
        self.hi_there["command"] = self.create_char
        self.hi_there.pack(side="top")

        self.quit = tk.Button(self, text="QUIT", fg="red",
                              command=self.master.destroy)
        self.quit.pack(side="bottom")

    def create_char(self):
        clear_output(wait=True)
        with urllib.request.urlopen("https://rickandmortyapi.com/api/character/") as url:
            data = json.loads(url.read().decode())
        characters_pool = data['results']
        character = characters_pool[random.randint(0,len(characters_pool)-1)]
        attributes = list(character.keys())
        for attribute in attributes[:6]:
            print(f'{attribute}: {character[attribute]}')
        url_image = character['image']
        im = Image.open(urlopen(url_image))
        plt.imshow(im)
        plt.show()
        print('--------')


root = tk.Tk()
app = Application(master=root)
app.mainloop()
```
